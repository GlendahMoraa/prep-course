# **Loading Data**


```python
#importing Pandas

import pandas as pd
```

```python
# loading dataset 1
dt1 = 'http://bit.ly/Telcom_dataset1'
data1 = pd.read_csv(dt1)
dataset1 = data1.set_axis(['PRODUCT', 'VALUE', 'DATETIME', 'CELL_ON_SITE', 'DW_A_NUMBER_INT', 'DW_B_NUMBER_INT', 'COUNTRY_A', 'COUNTRY_B', 'CELL_ID', 'SITE_ID'],  axis=1, inplace=False)
dataset1
```

```python
#loading Dataset2
dt2 = 'http://bit.ly/Telcom_dataset2'
data2 = pd.read_csv(dt2)
dataset2 = data2.set_axis(['PRODUCT', 'VALUE', 'DATETIME', 'CELL_ON_SITE', 'DW_A_NUMBER_INT', 'DW_B_NUMBER_INT', 'COUNTRY_A', 'COUNTRY_B', 'CELL_ID', 'SITE_ID'],  axis=1, inplace=False)
dataset2
```

```python
#loadind Dataset3 
dt3 = 'http://bit.ly/Telcom_dataset3'
data3 = pd.read_csv(dt3)
dataset3 = data3.set_axis(['PRODUCT', 'VALUE', 'DATETIME', 'CELL_ON_SITE', 'DW_A_NUMBER_INT', 'DW_B_NUMBER_INT', 'COUNTRY_A', 'COUNTRY_B', 'CELL_ID', 'SITE_ID'],  axis=1, inplace=False)
dataset3
```

```python
#loading Cells_geo dataset

cells = pd.read_csv('/content/cells_geo.csv', delimiter=';')
cells_geo = cells.set_axis([' 0', 'CITY', 'STATUS', 'LOCALISATION', 'DECOUPZONE', 'ZONENAME', 'LONGITUDE', 'LATITUDE', 'REGION', 'AREA', 'CELL_ID', 'SITE_ID'],  axis=1, inplace=False)
cells_geo
```

## 1.1 Merging DataFrames

```python
my_data = [dataset1, dataset2, dataset3]
mtn_telcom = pd.concat(my_data)
mtn_telcom
```

```python
#At this point we can now find the usage in the there days between office hours and home hours 

# Lets start by looking at working hours..
#

working_hrs = mtn_telcom[(mtn_telcom['DATETIME'].dt.hour >= 8) & (mtn_telcom['DATETIME'].dt.hour <= 17)] ["PRODUCT"].value_counts()
working_hrs

#from the output we realise that no product was used used during the buisness hours.
#so this means that its either people dont use their phones during the business hours or,
#the data collected was for home hours.
```

```python
#now we can look at home hours

after_work = mtn_telcom[(mtn_telcom['DATETIME'].dt.hour >= 17) & (mtn_telcom['DATETIME'].dt.hour <= 24)] ["PRODUCT"].value_counts()
before_work= mtn_telcom[(mtn_telcom['DATETIME'].dt.hour >= 0) & (mtn_telcom['DATETIME'].dt.hour < 8)] ["PRODUCT"].value_counts()


home_hours = after_work + before_work
home_hours

#we realise that each product was used during this hours which means
#1. the date collected may have been for the homehours only or in this cities people use there phones during home hours only
#2. the results also show that the most product that is used is sms so the company can thinkmof investing more on sms
```
