```python
import csv
import pandas as pd
```

```python
with open('Uganda_Karamoja_Subcounty_Crop_Yield_Population.csv', 'r') as f:
    Uganda_Karamoja_Subcounty_Crop_Yield_Population = pd.read_csv(f, index_col=0, encoding='utf-8')

%sql DROP TABLE IF EXISTS Uganda_Karamoja_Subcounty_Crop_Yield_Population;
%sql PERSIST Uganda_Karamoja_Subcounty_Crop_Yield_Population
```

```python
#Previewing the Uganda_Karamoja_Subcounty_Crop_Yield_Population table

%%sql
SELECT * FROM Uganda_Karamoja_Subcounty_Crop_Yield_Population limit 5;
```

```python
# Selecting sub counties their population and area

%%sql
SELECT "SUBCOUNTY_NAME", POP, Area FROM Uganda_Karamoja_Subcounty_Crop_Yield_Population
```

```python
# Sort the list of districts by total crop area (descending order).
%%sql
SELECT "DISTRICT_NAME", SUM("Crop_Area_Ha") FROM Uganda_Karamoja_Subcounty_Crop_Yield_Population
GROUP BY "DISTRICT_NAME"
ORDER BY "Crop_Area_Ha" DESC
```

```python
#select only subcounties from the moroto district order by aphabet and show there production in soghum

%%sql
SELECT "SUBCOUNTY_NAME", "DISTRICT_NAME", "S_Prod_Tot" FROM Uganda_Karamoja_Subcounty_Crop_Yield_Population
WHERE "DISTRICT_NAME" = 'MOROTO'
ORDER BY "SUBCOUNTY_NAME" ASC
```

```python
# Compute the total Maize production per District.
%%sql
SELECT "DISTRICT_NAME", SUM("M_Prod_Tot") FROM Uganda_Karamoja_Subcounty_Crop_Yield_Population
GROUP BY "DISTRICT_NAME"
```

```python
# Compute the number of Sub counties where Maize is produced and the total Maize production per District.
%%sql
SELECT COUNT("SUBCOUNTY_NAME"), "DISTRICT_NAME", "M_Yield_Ha", SUM("M_Prod_Tot") FROM Uganda_Karamoja_Subcounty_Crop_Yield_Population
WHERE "M_Yield_Ha" > 0
GROUP BY "DISTRICT_NAME"
```

```python
# Compute the overall Crop area in all Sub counties where population is over 20000.
%%sql
SELECT "SUBCOUNTY_NAME", "Crop_Area_Ha", POP FROM Uganda_Karamoja_Subcounty_Crop_Yield_Population
WHERE POP > 20000
```

```python
# Sort the Maize production in descending order by Districts, 
#only taking into account Sub counties where Maize area is larger than Sorghum area, and 
#display the number of Sub counties per district matching that criteria.
%%sql
SELECT COUNT("SUBCOUNTY_NAME"), "DISTRICT_NAME", "M_Prod_Tot" FROM Uganda_Karamoja_Subcounty_Crop_Yield_Population 
WHERE "M_Yield_Ha" > "S_Yield_Ha"
GROUP BY "DISTRICT_NAME"
ORDER BY "M_Prod_Tot" DESC
```
