## 1.4 Answering Questions

Let's answer the following questions with our dataset

```
# Challenge 1
# How many trips were taken on the 22/04/18?
# 
df
df.loc[df['travel_date'] == '2018-04-22'].value_counts()
```

```
# Challenge 2
# Which type of vehicle travelled on 23/04/18 from Kisii to Nairobi at 7.15am?
#
df.loc[(df['travel_date'] == '2018-04-23') & (df['travel_time'] == '07:15') & (df['travel_from'] == 'Kisii') & (df['travel_to'] == 'Nairobi')]['car_type']
```

```
# Challenge 3
# How many trips were made from Kisii to Nairobi?
# 
df.loc[(df['travel_from'] == 'Kisii') & (df['travel_to'] == 'Nairobi')]['ride_id'].count()
```

```
# Challenge 4
# How many trips were made by shuttle from Nairobi?
# 
df.loc[(df['travel_from'] == 'Nairobi') & (df['car_type'] == 'shuttle')]['ride_id'].count()
```

```
# Challenge 5
# How many trips were made by bus from Nairobi to Awendo?
# 
df.loc[(df['travel_to'] == 'Awendo') & (df['travel_from'] == 'Nairobi') & (df['car_type'] == 'Bus')]['ride_id'].count()
```

```
# Challenge 6
# Which was the most travelled from destination by bus?
# 
most_travelled = df.loc[df['car_type'] == 'Bus']['travel_from'].value_counts()
most_travelled.sort_values(ascending= False).head(1)
```

```
# Challenge 12
# If all shuttles travelled with full capacity, how many people travelled in shuttles from Kisii to Nairobi in the month of May?
#
trips = df.loc[(df['date'].dt.month == 5) & (df['travel_from'] == 'Kisii') & (df['car_type'] == 'shuttle') & (df['travel_to'] == 'Nairobi')]['ride_id'].count()
total_people = trips * 11
total_people
```
